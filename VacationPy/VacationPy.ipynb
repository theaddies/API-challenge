### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
hotel_df['Hotel Name'] = 0
```

```python
hotel_df = hotel_df.rename(columns = {'city' : 'City', 'Latitude': 'Lat', 'Longitude': 'Lng'})
```

```python
for city in hotel_df['City']:
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

    target_type = "lodging"
    lat = hotel_df.loc[hotel_df['City'] == city , 'Lat'].to_numpy()[0]
    lon = hotel_df.loc[hotel_df['City'] == city , 'Lng'].to_numpy()[0]
    coords = str(lat) + " , " + str(lon)
    print('coordinates = ',coords)
    #coords = "47.6062095,-122.3320708"
    radius = 5000
    params = {
        "location": coords,
        "types": target_type,
        "radius": radius,
        "key": g_key
    }

    # Build URL using the Google Maps API
    try:
        response= requests.get(base_url, params).json()['results'][0]['name']
    except IndexError:
        response = "no hotel"
    hotel_df.loc[hotel_df['City'] == city,'Hotel Name'] = response
hotel_df
```
